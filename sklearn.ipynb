```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
iris = datasets.load_iris()
```

```python
df_X = pd.DataFrame(iris.data)
df_y = pd.DataFrame(iris.target)
#df_X.head()
#df_y.head()
```

train_test_split根本神器阿！
機器學習會將資料分成訓練資料及測試資料
監督學習則會有標籤(label)，可以簡單的將資料切分
1. train代表訓練集 test代表測試集
2. X代表train data Y代表train label
3. 顯示後可發現data使經過洗牌(shuffle)過的

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.3)
```

選擇機器學習的模型，調整參數
這裡使用決策樹做分類 

DecisionTreeClassifier()參數說明
1. criterion: gini v.s. entropy
2. splitter: best vs random
3. class_weight={'類別':權重(int)}
理解：
1. 速度: gini > entropy
2. splitter: 大資料建議random
3. class_weight: 資料歪斜時可調整，少的資料權重高

```python
clf = tree.DecisionTreeClassifier()
iris_clf = clf.fit(X_train, y_train)
```

評估模型

```python
test_y_predicted = iris_clf.predict(X_test)
```

```python
accuracy = metrics.accuracy_score(y_test, test_y_predicted)
print(accuracy)
```
